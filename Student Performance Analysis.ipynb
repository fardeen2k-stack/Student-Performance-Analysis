checking missing values 

```python
print(df.isnull().sum())
```

Find Average Score 

```python
print("Math Avg:", df["math score"].mean())
print("reading Avg:", df["reading score"].mean())
print("Writing Avg:", df["writing score"].mean())
```

find top Student 

```python
top_student = df[df["math score"] > 90 ]
print(top_student)
```

Compare Performance by Gender

```python
print(df.groupby("gender")[["math score", "reading score", "writing score"]].mean())
```

totl score column 

```python
df["total score"] = df["math score"] + df["reading score"] + df[ "writing score"]
print(df.head())
```

find best student 

```python
best =df.sort_values(by ="total score", ascending= False )
print(best.head(1))
```

```python
import matplotlib.pyplot as plt
df[["math score","reading score","writing score"]].hist()
```

```python
df["math score"].hist()
plt.title("Math Score Distribution ")
plt.xlabel("Math Score")
plt.ylabel("Number of Student")
plt.show()
```

Bar plot & Heat map

```python
import seaborn as sns 

sns.barplot(x="gender", y="math score", data=df)
plt.title("Average Math Score by Gender")
```

```python
corr = df[["math score","reading score","writing score"]].corr()
sns.heatmap(corr, annot=True)
plt.title("Relationship Between Student Performance Metrics")
```
